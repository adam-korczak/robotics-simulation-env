## Matplotlib basics

```python
import random
import matplotlib.pyplot as plt

# Gaussian distribution
data = [random.gauss(0, 10) for i in range(1000)]
plt.hist(data, bins=10, range=(-30, 30))
plt.xlabel('Value of Sample')
plt.ylabel('Number of Samples')
plt.title('This is a histogram')
plt.show()
```

## Plot square

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_square(x, y, side_length):
    """
    Plots a square using matplotlib.

    Parameters:
    - x: The x-coordinate of the bottom-left corner of the square.
    - y: The y-coordinate of the bottom-left corner of the square.
    - side_length: The length of a side of the square.
    """
    
    fig, ax = plt.subplots()
    square = patches.Rectangle((x, y), side_length, side_length, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(square)

    # Set the axes limits to ensure the square is visible
    ax.set_xlim(x - side_length/2, x + 1.5*side_length)
    ax.set_ylim(y - side_length/2, y + 1.5*side_length)
    ax.set_aspect('equal', 'box')  # This ensures the square looks like a square

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    plt.title('Square Plot')
    plt.show()

# Example usage:
plot_square(0, 0, 0.2)

```
